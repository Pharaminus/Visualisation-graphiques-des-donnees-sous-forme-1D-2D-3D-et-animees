### **VISUALISATION 1D D'UN JEU DE DONNEE( histogramme, densite et boite a moustache)**

### **Pratique 1: Histogramme et densitÃ©**

### Pour la production de ces deux graphiques nous allons implementer une classe visualisation 1D qui vas gerer l'affichage des graphes 1D. Nous allons utiliser un jeu de donnees generer aleatoirement suivant une loi normale mais la classe sera adapter pour recevoir les parametres necessaire a l'affichage des graphiques.

#### **Importation des bibliotheques**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#### **Generation ou importation des donnees**.
#### vous pouvez importez vos donnees dans le bloque suivant.

```python
n_sample = 1000
mean = 0                #====|> moyenne des donnees
stddev = 1              #===|> Ecartype des donnees
outlier_ratio = 0.7     #===|> Proportion des outliers
outlier_mean = 4       #===|> moyenne des outliers
outlier_stdev = 5       #===|> Ecartype des outliers
n_outliers = 80         #===|> nombre d'outliers

data = np.random.normal(mean, stddev, n_sample)
outlier = np.random.normal(outlier_mean, outlier_stdev, n_outliers)
# data = np.random.normal(mean, outlier_stdev, int(n_sample * outlier_ratio))

data = np.concatenate((data, outlier))
```

#### **Methode pour la production graphique 1D**

```python
# =======( Histogramme et densitÃ© )========

plt.figure(figsize = (9, 5))
sns.histplot(data, bins=100, kde=True, color='blue')
plt.title("Histograme et densitÃ©")
plt.xlabel("Valeurs")
plt.ylabel("Frequences")
plt.show()
```

### **Pratique 2: Boxplot et Outliers**

Nous generons un jeu de donnees contenant des outliers.

```python
n_sample = 1000
mean = 0                #====|> moyenne des donnees
stddev = 1              #===|> Ecartype des donnees
outlier_ratio = 0.7     #===|> Proportion des outliers
outlier_mean = 2       #===|> moyenne des outliers
outlier_stdev = 5       #===|> Ecartype des outliers
n_outliers = 3         #===|> nombre d'outliers

data = np.random.normal(mean, stddev, n_sample)
outlier = np.random.normal(outlier_mean, outlier_stdev, n_outliers)
# data = np.random.normal(mean, outlier_stdev, int(n_sample * outlier_ratio))

data = np.concatenate((data, outlier))
```

```python
# ======( Boxplot et Outliers )======
sns.boxplot(data, orient='h', color='b')
# plt.boxplot(data, vert=False, patch_artist=True)
plt.title("Boite a moustache des donnees")
plt.show()
```
